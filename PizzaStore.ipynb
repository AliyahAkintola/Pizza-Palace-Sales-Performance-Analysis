## KPIs

#### Total Revenue

```python
%%sql
select CAST(SUM(total_price) as decimal(10,2)) as total_revenue
from pizzastore.pizza_sales;
```

#### Average Order Value

```python
%%sql
select CAST(SUM(total_price)/COUNT(DISTINCT(order_id)) AS decimal(10,2)) AS Avg_Order_Value
from  pizzastore.pizza_sales;
```

#### Total Pizzas Sold

```python
%%sql
SELECT SUM(quantity) as Total_pizzas_sold
from  pizzastore.pizza_sales;
```

#### Total Orders

```python
%%sql
select COUNT(DISTINCT(order_id)) as Total_orders
from pizzastore.pizza_sales;
```

#### Average Pizzas Per Order

```python
%%sql 
select CAST(SUM(quantity)/COUNT(DISTINCT(order_id)) AS decimal(10,2)) as Avg_pizzas_per_order
from pizzastore.pizza_sales;
```

## Chart Requirements

#### Daily Trend for Total Orders

```python
%%sql
select 
     DAYNAME(order_date) as Order_day,
     COUNT(DISTINCT(order_id)) as Total_orders
from  pizzastore.pizza_sales
group by WEEKDAY(order_date),dayname(order_date)
order by weekday(order_date) ASC;
```

#### Monthly Trend for Total Orders

```python
%%sql
select 
     monthname(order_date) as Month_name,
     COUNT(DISTINCT(order_id)) as Total_orders
from  pizzastore.pizza_sales
group by monthname(order_date),month(order_date)
order by month(order_date);
```

#### Percentage of Sales by Pizza Category

```python
%%sql
select
pizza_category,
count(quantity) as Qty_sold_per_category,
cast(sum(total_price) as decimal(10,2)) as sales_per_caategory,
cast(sum(total_price)*100/(select sum(total_price) from pizza_sales) as decimal(10,2)) Percentage_of_total_sales
from pizza_sales
group by pizza_category;
```

#### Percentage of Sales by Pizza Size

```python
%%sql
select
pizza_size,
count(quantity) as qty_sold_per_size,
cast(sum(total_price) as decimal(10,2)) as sales_per_size,
cast(sum(total_price)*100/(select sum(total_price) from pizza_sales) as decimal(10,2)) Percentage_of_total_sales
from pizza_sales
group by pizza_size
order by Percentage_of_total_sales DESC;
```

#### Best Performing Products by revenue

```python
%%sql
select pizza_name,
cast(sum(total_price) as decimal(10,2)) as Total_revenue
from pizza_sales
group by pizza_name
order by Total_revenue desc
limit 5;
```

#### Least Performing Products by revenue

```python
%%sql
select pizza_name,
cast(sum(total_price) as decimal(10,2)) as Total_revenue
from pizza_sales
group by pizza_name
order by Total_revenue asc
limit 5;
```

#### Best Performing Products by quantity

```python
%%sql
select pizza_name,
sum(quantity)as Qty_sold
from pizza_sales
group by pizza_name
order by Qty_sold desc
limit 5;
```

#### Least Performing Products by Quantity

```python
%%sql
select pizza_name,
sum(quantity)  as Qty_sold
from pizza_sales
group by pizza_name
order by Qty_sold asc
limit 5;
```

#### Best Performing Products by Orders

```python
%%sql
select pizza_name,
count(distinct(order_id)) as Total_orders
from pizza_sales
group by pizza_name
order by Total_orders desc
limit 5;
```
